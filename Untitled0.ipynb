```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


x=[5,10,15,20,25]
y=[6,12,18,24,30]
plt.plot(x,y, color ='blue')

plt.savefig("erfr.png")
```

```python
plt.plot(x,y,color="blue",marker='h',markersize=20,markeredgecolor="black")
plt.xlabel(x)
plt.ylabel(y)
```

```python
x=["python", "c","c++","java"]
y=[85,52,60,45]
plt.xlabel("Language",fontsize=10)
plt.ylabel("Number")
plt.title("Result",fontsize=20)
plt.bar(x,y,color="blue",width=0.5)
plt.show(x,y)
```

```python
c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color=c,alpha=0.5,label="language")
plt.legend()
plt.show()
```

```python
fruits=["apple","mango","banana","orange"]
values=[30,25,20,15]
ex=[0.2,0.3,0.0,0.0]


plt.pie(values, labels=fruits,autopct="%0.1f%%",explode=ex)

plt.title("pie chart ")
plt.legend(loc=2)
plt.show()
```
